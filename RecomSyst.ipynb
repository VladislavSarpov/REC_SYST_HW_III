```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
movies_with_ratings = pd.merge(ratings, movies)
movies_with_ratings
```

Строим матрицу данных с оценками пользователей:

```python
movie_ratings = movies_with_ratings.pivot_table(index=['userId'], columns=['title'], values='rating')

movie_ratings
```

Эта строка для того, что бы правильно записать название фильма (можно не использовать) :) 

```python
[movie for movie in movie_ratings.columns if 'Star Wars' in movie]
```

выбираем фильм и генерируем индекс схожести фильмов:

```python
star_wars_ratings = movie_ratings['Star Wars: Episode IV - A New Hope (1977)']

similar_movies = movie_ratings.corrwith(star_wars_ratings)
similar_movies = similar_movies.dropna()
df = pd.DataFrame(similar_movies)

df
```

Удаляем фильмы с низким рейтингом 

```python
ratings_count = 100
movie_stats = movies_with_ratings.groupby('title').agg({'rating': [np.size, np.mean]})['rating']
popular_movies = movie_stats['size'] >= ratings_count
movie_stats[popular_movies].sort_values('mean', ascending=False)[:15]
```

Отбираем схожие фильмы, имеющие высокий рейтинг:

```python
df = movie_stats[popular_movies].join(similar_movies.to_frame(name='similarity'))
df.sort_values(['similarity'], ascending=False)[:15]
```

Подбираем фильмы на основе оценок выставленных пользователем (например, UserId ==2, он выставил 29 оценок)

```python
corr_matrix = movie_ratings.corr(method='pearson', min_periods=100)
corr_matrix
```

```python
for col in corr_matrix.columns:
    values = corr_matrix[col].unique()
    if len(values) > 1:
        print(col, corr_matrix[col].unique())
```

```python
movies_with_ratings[movies_with_ratings['userId'] == 2]
```

```python
my_ratings = movie_ratings.loc[2].dropna()
sim_candidates = pd.Series()
for i in range(0, len(my_ratings.index)):
    # Извлекаем фильмы, похожие на оцененные мной
    sims = corr_matrix[my_ratings.index[i]].dropna()
    # Далее оцениваем их сходство в зависимости от того, как я оценил тот или иной фильм
    sims = sims.map(lambda x: x * my_ratings[i])
    # Добавляем индекс в список сравниваемых кандидатов
    sim_candidates = sim_candidates.append(sims)
    
sim_candidates.sort_values(inplace = True, ascending = False)

sim_candidates
```

```python
sim_candidates = sim_candidates.groupby(sim_candidates.index).sum()
sim_candidates.sort_values(inplace = True, ascending = False)

sim_candidates
```

```python
filtered_sims = sim_candidates.drop(my_ratings.index, errors='ignore')
filtered_sims[:10]
```
